```python
import numpy
import seaborn as sns
from bootstrapper import Bootstrapper
```

```python
numpy.random.seed = 20200209
sample = numpy.random.power(0.1, 10_000)
b = Bootstrapper(bootstrap_count=1_000)
low, high, bs_samples = b.ci(lambda array: numpy.mean(array), 0.95, sample)
```

```python
sns.set()
ax = sns.distplot(bs_samples)
ax.axvline(low, color='red')
ax.axvline(high, color='red');
```
