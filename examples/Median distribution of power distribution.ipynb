# Import

```python
import numpy
import seaborn as sns
from bootstrapper import Bootstrapper
```

# Bootstrapping
1. Generate random numbers following power distribution
2. Setup a Bootstrapper instance
3. Feed generated numbers to Bootstrapper to get the distribution of median.

```python
numpy.random.seed = 20200122
sample = numpy.random.power(0.1, 10_000)
b = Bootstrapper(bootstrap_count=1_000)
quantiles = b.run(lambda array: numpy.median(array), sample)
```

# Plotting

```python
sns.set()
sns.distplot(quantiles, kde=True);
```
